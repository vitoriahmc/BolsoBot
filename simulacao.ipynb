```python
import random
import numpy as np
TIMES = 100000

portas = [(2, True), (3, False), (5, False)]
h = []
for i in range(TIMES):
    horas = 0
    escaped = False
    while not escaped:
        new_horas, escaped = random.choice(portas)
        horas += new_horas
    h.append(horas)
a = np.array(h)
print(f'MEAN: {a.mean()}')
print(f'MAX: {a.max()}')
print(f'MIN: {a.min()}')
```

```python
TIMES = 1000000
p = []
for i in range(TIMES):
    success = False
    paid = 0
    t = 0
    while not success:
        if t == 0:
            paid += 1000
        elif t > 0 and t <= 3:
            paid += 400
        else:
            paid += 200
        chance = random.random()
        if chance <= 0.2:
            success = True
        t+=1
    p.append(paid)
p = np.array(p)
print(f'MEAN: {p.mean()}')
print(f'MAX: {p.max()}')
print(f'MIN: {p.min()}')
```
